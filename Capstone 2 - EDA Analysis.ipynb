# Cost of Living EDA Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

df = pd.read_csv('costofliving.csv')
```

```python
print(df)
```

```python
df.head()
```

```python
df.set_index('Number')
```

```python
df.describe()
```

```python
sns.heatmap(df.drop('Number',axis=1).corr(),vmin=0.0)
```

```python
sns.pairplot(df.drop('Number',axis=1),corner =True  )
```

**Analysis**

-These graphs help to explain the correlations between the cost of living, rent, and both combined indices. Since most of the features have a positive correlation, it proves to show that each indice has a positive correlation to one another. The only Index that doesnt have that strong positive correlation is the Local Purchasing Power Index. The reasoning behind this will be observed over the next few analysis, however, this can be comprosomised due to the supply and demand of products in each respective City

```python
df = df.drop(labels=['Cost of Living Index','Rent Index'],axis=1)
df.head()
```

```python
df['total_price_index'] = df[['Cost of Living Plus Rent Index','Groceries Index','Restaurant Price Index']].sum(axis=1)/3
```

### Most Expensive Cities to Live in

```python
expensive = df.sort_values(by='total_price_index',ascending=False).head(20)
expensive
```

```python
plt.figure(figsize=(10,10))
sns.barplot(x=expensive['total_price_index'],y=expensive['City'])
```

```python
plt.figure(figsize=(10,10))
sns.barplot(x=expensive['total_price_index'],y=expensive['State'])
```

```python
plt.figure(figsize=(10,10))
sns.barplot(x=expensive['Local Purchasing Power Index'],y=expensive['City'])
```

```python
plt.figure(figsize=(10,10))
sns.barplot(x=expensive['Local Purchasing Power Index'],y=expensive['State'])
```

### Cheapest City to live in

```python
cheap = df.sort_values(by='total_price_index').head(20)
cheap
```

```python
fig,axes = plt.subplots(1,2,figsize=(10,10))
sns.barplot(x=cheap['Local Purchasing Power Index'],y=cheap['City'],ax=axes[0])
sns.barplot(x=cheap['total_price_index'],y=cheap['City'],ax=axes[1])

fig,axes = plt.subplots(1,2,figsize=(10,10))
sns.barplot(x=cheap['Local Purchasing Power Index'],y=cheap['State'],ax=axes[0])
sns.barplot(x=cheap['total_price_index'],y=cheap['State'],ax=axes[1])
```

### Cities with the Highest Purchasing Power


```python
purchase = df.sort_values(by='Local Purchasing Power Index',ascending =False).head(20)
purchase
```

```python
plt.figure(figsize=(10,10))
sns.barplot(x=purchase['Local Purchasing Power Index'],y=purchase['City'])

plt.figure(figsize=(10,10))
sns.barplot(x=purchase['Local Purchasing Power Index'],y=purchase['State'])
```

After the analysis of observing Highest purchasing power cities, we can notice that Bellevue, WA and San Jose, CA have the highest local purchasing powers. Respectively, we can also analyze that WA and CA are the 2 highest rankings for the local purchasing power indicators.
